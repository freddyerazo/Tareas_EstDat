```python
from dataclasses import dataclass
import re
```

```python
##INSTANCIA DEL REGISTRO
@dataclass
class Consultor:
    id: int
    consultor: str
    area: str
    horas: int
    salario_semanal: float
```

```python
##data set completo
data_set = [
    "ID:001; NAME: alberto_EINSTEIN ; PROYECTO:Relatividad; HORAS:40; PAGO_H:35.5",
    "id-002, name- marie_CURIE, proyecto-Radiactividad, horas-35, pago_h-40.0",
    "ID:003; NAME:isaac_NEWTON; PROYECTO:Leyes del Movimiento; HORAS:20; PAGO_H:30.0####",
    " id:004 - name: nikola_TESLA - proyecto:Corriente Alterna - horas:50 - pago_h:25.0",
    "ID:005; NAME:ada_LOVELACE; PROYECTO:Algoritmos; HORAS:45; PAGO_H:45.0",
    "id-006, name- charles_DARWIN, proyecto-Evolución, horas-30, pago_h-20.0????",
    "ID:007; NAME: rosalind_FRANKLIN ; PROYECTO:Estructura ADN; HORAS:38; PAGO_H:32.5",
    "id:008 - name: stephen_HAWKING - proyecto:Agujeros Negros - horas:15 - pago_h:50.0",
    "ID:009; NAME:katherine_JOHNSON; PROYECTO:Trayectorias NASA; HORAS:42; PAGO_H:40.0!!!!",
    "id-010, name- grace_HOPPER, proyecto-Compiladores, horas-48, pago_h-38.0"
]
```

```python
lista_consultores = []
```

```python

for linea in data_set:

    # - 1: Eliminar (#, ?, !)
    linea = re.sub(r'[#?!]+', '', linea).strip()
    # - 2: Normalizar " - " con espacios a coma (separador de campos)
    linea = re.sub(r'\s+-\s+', ',', linea)
    # - 3: Normalizar ";" a coma
    linea = re.sub(r'[;]', ',', linea)
    # - 4: Split por coma para obtener campos
    partes = linea.split(',')
    # - 5: Extraer clave: valor de cada parte
    registro = {}
    for parte in partes:
        parte = parte.strip()
        if ':' in parte:
            clave, valor = parte.split(':', 1)
            registro[clave.strip().upper()] = valor.strip()
        elif '-' in parte:
            clave, valor = parte.split('-', 1)
            registro[clave.strip().upper()] = valor.strip()

    # - 6: Construir datos
    nombre  = registro.get('NAME', '').replace('_', ' ').strip().title()
    area    = registro.get('PROYECTO', '').strip().upper()
    horas   = int(registro.get('HORAS', 0))
    pago_h  = float(registro.get('PAGO_H', 0))
    salario = round(horas * pago_h, 2)

    
    lista_consultores.append(Consultor(
        id=int(registro.get('ID', 0)),
        consultor=nombre,
        area=area,
        horas=horas,
        salario_semanal=salario
    ))
    

```

```python
# MOSTRAR RESULTADO
print(f"{'ID':<5} {'CONSULTOR':<28} {'AREA':<30} {'SALARIO A RECIBIR'}")
print("-" * 85)
for inv in lista_consultores:
    print(f"{inv.id:<5} {inv.consultor:<28} {inv.area:<30} ${inv.salario_semanal}")
```

```python
# PARTE 2 - Mayor salario

# Consultor con mayor salario semanal
consultor_max = max(lista_consultores, key=lambda c: c.salario_semanal)
print(f"El investigador con el salario mas alto es: {consultor_max.consultor}")
print(f"Salario semanal: ${consultor_max.salario_semanal}")
```

```python
# PARTE 2 - Promedio de horas trabajadas de todo el listado
promedio_horas = sum(c.horas for c in lista_consultores) / len(lista_consultores)
print(f"Promedio de horas trabajadas: {promedio_horas:.1f} horas")
```

```python
# PARTE 3 - Generador de credenciales de acceso SISTEMA INTERNO
# Logica: CONS- + 3 letras apellido + ID con ceros

print(f"\n{'CONSULTOR':<28} {'ACCESS-USER'}")
print("-" * 55)
for c in lista_consultores:
    apellido   = c.consultor.split()[-1][:3].upper()
    credencial = f"CONS-{apellido}{str(c.id).zfill(3)}"
    print(f"{c.consultor:<28} {credencial}")
```

PREGUNTA 1:

Si un registro no tuviera el campo 'PAGO_H', ¿cómo reaccionaría su código al intentar calcular el salario? ¿Qué función de cadenas usarían para verificar si el campo existe antes de operar?

Si el registro no trae el campo, el código básicamente "se cae". Al intentar forzar un float() o hacer una multiplicación con algo que no existe, Python nos lanzaría un ValueError o un TypeError y detendría todo el proceso.
Para no complicarnos, lo ideal es usar el método .get(). En lugar de llamar a la llave directamente, usamos registro.get('PAGO_H', 0). Con esto, si el campo falta, el sistema simplemente asume que es 0 y sigue adelante sin lanzar ningún error ni detener la ejecución.

PREGUNTA 2:

¿Cuál fue el método más eficiente que encontraron para manejar tres tipos diferentes de separadores (;, ,, -)? ¿Usaron múltiples. replace() o una estrategia distinta?

Manejar tres tipos de separadores con puros .replace() es buscarse un dolor de cabeza. Por eso, nos fuimos por el camino más eficiente: Expresiones Regulares (re.sub).
Lo que realice es:
Normalización: Primero limpiamos cualquier caracter especial raro que sobrara.
El dilema del guion: Teníamos que ser cuidadosos. Usamos \s+-\s+ para atrapar solo los guiones que separan campos (los que tienen espacios a los lados) y convertirlos en comas. Si hubiéramos reemplazado todos los guiones, habríamos borrado los que separan las claves de los valores, como en pago_h-40.0.
Cierre: Finalmente, pasamos los puntos y coma (;) a comas. Así dejamos todo el dataset estandarizado y listo para trabajar.

PREGUNTA 3:

¿Si el dataset tuviera 1 millón de registros en lugar de 10, ¿creen que su código seguiría siendo rápido? ¿Qué parte del proceso de limpieza consume más memoria?

El código actual es robusto y seguiría procesando los datos, pero aquí el problema ya no es la lógica, sino la memoria RAM.
Ahora mismo estamos guardando todos los objetos Consultor en una lista. Con 10 datos es imperceptible, pero con un millón, esa lista se vuelve pesadísima y podría colgar la computadora.
Para que sea realmente rápido y eficiente, cambiaríamos la estrategia:
Librerías de alto rendimiento: Meterle Pandas o Polars, que están hechas justamente para "tragar" millones de filas sin pestañear.
En lugar de mostrar todo en pantalla al final, iríamos guardando los resultados directamente en un CSV o una base de datos conforme se van limpiando.
